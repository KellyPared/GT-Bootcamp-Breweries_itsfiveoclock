```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
meadery_df = pd.read_csv("data/untapped_meadery.csv")
micro_df = pd.read_csv("data/untapped_micros.csv")
micro_us_df = pd.read_csv("data/untapped_micros_us_brewery.csv")
nano_df = pd.read_csv("data/untapped_nano.csv")
regional_df = pd.read_csv("data/untapped_regional.csv")
```

```python
#meadery_df.head()
meadery_df[['number_beers', 'number_ratings']] = meadery_df['number_beers'].str.split('Beers ',1, expand=True)
meadery_df.head()
```

```python
meadery_df_clean = meadery_df[['brewery_name', 'brewery_location', 'brewery_type', 'number_beers', 'number_ratings', 'brewery_ratings']]
meadery_df_clean.head()
```

```python
meadery_df_clean['number_ratings'] = meadery_df_clean['number_ratings'].str.replace(r'[^0-9]', '')
meadery_df_clean['number_beers'] = meadery_df_clean['number_beers'].str.replace(r'[^0-9]', '')
meadery_df_clean['brewery_ratings'] = meadery_df_clean['brewery_ratings'].str.replace('(', '')
meadery_df_clean['brewery_ratings'] = meadery_df_clean['brewery_ratings'].str.replace(')', '')
#df["A"].str.replace("a","")
meadery_df_clean.head()
```

```python
meadery_df_clean.info()
```

```python
meadery_df_clean['number_beers'] = meadery_df_clean['number_beers'].astype(int)
meadery_df_clean['number_ratings'] = meadery_df_clean['number_ratings'].astype(int)
meadery_df_clean['brewery_ratings'] = meadery_df_clean['brewery_ratings'].astype(float)
```

```python
meadery_df_clean.info()
```

```python
meadery_df_clean.to_csv('cleaned_csvs/untappd_meadery_clean.csv', index=False)
```
